```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister
from qiskit.quantum_info import Statevector
```

```python
from qiskit.circuit.library.arithmetic.adders.draper_qft_adder import DraperQFTAdder
draper_adder = DraperQFTAdder(4)
draper_adder.decompose().draw("mpl")
```

```python
# Big endian
a, b = "0110", "1011"
state = Statevector.from_label(a + b)
state = state.evolve(draper_adder)
print(state[int("0001"+b, 2)])
```

```python
a, b = "0010", "0110"
state = Statevector.from_label(a + b)
state = state.evolve(draper_adder)
print(state[int("1000"+b, 2)])
```
